<b>Adding new Column having total count of documennt from 2 to 21 and dropped the flag document for others</b>

```python
######################################################
#Function Name :- normalizing_and_dropping_columns
#Input Parameters :- dataframe :- Dataframe that you want to work upon
#                    column_list :- list of columns that you want to drop
#                    action  :- action need to be performed for new column
#                    new_column_name :- Name of new_column
#
#It will take input of df , column list, action and new column name and will perform the action and add value to new column.
#After that it will drop those columns
#######################################################

def normalizing_and_dropping_columns(dataframe, column_list, action, new_column_name):
    
    if action == 'mean':
        dataframe[new_column_name] = dataframe[column_list].mean(axis=1)
        dataframe.drop(column_list, axis= 1, inplace = True)
        
    elif action == 'sum':
        dataframe[new_column_name] = dataframe[column_list].sum(axis=1)
        dataframe.drop(column_list, axis= 1, inplace = True)
        
    return dataframe
        
```

```python
doucment_column_list = ['FLAG_DOCUMENT_2',
       'FLAG_DOCUMENT_3', 'FLAG_DOCUMENT_4', 'FLAG_DOCUMENT_5',
       'FLAG_DOCUMENT_6', 'FLAG_DOCUMENT_7', 'FLAG_DOCUMENT_8',
       'FLAG_DOCUMENT_9', 'FLAG_DOCUMENT_10', 'FLAG_DOCUMENT_11',
       'FLAG_DOCUMENT_12', 'FLAG_DOCUMENT_13', 'FLAG_DOCUMENT_14',
       'FLAG_DOCUMENT_15', 'FLAG_DOCUMENT_16', 'FLAG_DOCUMENT_17',
       'FLAG_DOCUMENT_18', 'FLAG_DOCUMENT_19', 'FLAG_DOCUMENT_20',
       'FLAG_DOCUMENT_21']

application_data_after_removing_columns_df = normalizing_and_dropping_columns(application_data_after_removing_columns_df, doucment_column_list,'sum','FLAG_DOCUMENT_COUNT')
```
